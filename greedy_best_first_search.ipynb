```python
import matplotlib.pyplot as plt
import numpy as np

# Ukuran grid
cols, rows = 7, 4
grid = np.zeros((rows, cols))

# Tandai obstacle (1 = dinding)
obstacles = [(1,1), (1,2), (2,1), (3,1), (3,2), (5,0),(5,1),(5,3)]
for (x, y) in obstacles:
    grid[y, x] = 1

# Node dan posisi koordinat (x, y)
nodes = {
    'A': (0,0), 'B': (0,3), 'C': (2,3), 'D': (2,2),
    'E': (4,3), 'F': (4,2), 'G': (4,0), 'H': (6,2),
    'I': (6,3), 'J': (6,0)
}

# Jalur hasil GBFS dari perhitungan manual
path = ['A', 'G', 'F', 'H', 'J']

# Gambar grid
plt.figure(figsize=(8,5))
plt.imshow(grid, cmap='Greys', origin='upper')

# Gambar node
for n, (x, y) in nodes.items():
    plt.scatter(x, y, s=500, c='lightblue', edgecolors='black', zorder=3)
    plt.text(x, y, n, ha='center', va='center', fontsize=12, fontweight='bold')

# Gambar jalur (warna merah)
for i in range(len(path)-1):
    x1, y1 = nodes[path[i]]
    x2, y2 = nodes[path[i+1]]
    plt.plot([x1, x2], [y1, y2], 'r-', linewidth=3, zorder=2)

# Tampilkan gridlines
plt.xticks(range(cols))
plt.yticks(range(rows))
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Representasi Grid dengan Obstacle dan Jalur Greedy Best First Search", fontsize=13)
plt.gca().invert_yaxis()
plt.show()
```
