```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/refs/heads/main/dataset/titanic.csv"
df=pd.read_csv(path)
df.head()
```

```python
df.isna().sum()
```

```python
drop_columns=['PassengerId','Name','Ticket','Cabin']
df.drop(columns=drop_columns, inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.Embarked.value_counts()
```

```python
df.Age.fillna(df.Age.mean(), inplace=True)
df.Embarked.fillna('S',inplace=True)
df.isna().sum()
```

```python
target='Survived'
y=df[target]
X=df.drop(columns=[target])
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
X['Sex']=le.fit_transform(X.Sex)

X.head()
```

```python
X['Embarked']=le.fit_transform(X.Embarked)
X.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,y_train,x_test,y_test=train_test_split(X,y, test_size=0.2, shuffle=True, stratify=y, random_state=100)
```
